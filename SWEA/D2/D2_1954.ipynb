# 1954. 달팽이 숫자

```

다음과 같이 정수 N을 입력 받아 N크기의 달팽이를 출력하시오.

[제약사항]
달팽이의 크기 N은 1 이상 10 이하의 정수이다. (1 ≤ N ≤ 10)

[입력]
가장 첫 줄에는 테스트 케이스의 개수 T가 주어지고, 그 아래로 각 테스트 케이스가 주어진다.
각 테스트 케이스에는 N이 주어진다.

[출력]
각 줄은 '#t'로 시작하고, 다음 줄부터 빈칸을 사이에 두고 달팽이 숫자를 출력한다.
(t는 테스트 케이스의 번호를 의미하며 1부터 시작한다.)
```

````python
#런타임 에러발생.. numpy를 못불러오는건가?
import numpy as np
T = int(input())

output = []

for k in range(T):
    N = int(input())
    result = np.zeros((N,N))
    x = N//2
    y = (N-1)//2
    cnt = N**2
    z = 1

    result[x][y] = cnt

    if N%2:
        while z<N:

            #홀수 , 왼쪽 + 아래
            if z%2:
                for i in range(z):
                    cnt -= 1
                    y -= 1
                    result[x][y] = cnt
                for i in range(z):
                    cnt -= 1
                    x += 1
                    result[x][y] = cnt
            #짝수, 오른쪽 + 위
            else:
                for i in range(z):
                    cnt -= 1
                    y += 1
                    result[x][y] = cnt
                for i in range(z):
                    cnt -= 1
                    x -= 1
                    result[x][y] = cnt
            z +=1
    else:
        while z<N:
            #짝수 , 왼쪽 + 아래
            if z%2==0:
                for i in range(z):
                    cnt -= 1
                    y -= 1
                    result[x][y] = cnt
                for i in range(z):
                    cnt -= 1
                    x += 1
                    result[x][y] = cnt
            #홀수, 오른쪽 + 위
            else:
                for i in range(z):
                    cnt -= 1
                    y += 1
                    result[x][y] = cnt
                for i in range(z):
                    cnt -= 1
                    x -= 1
                    result[x][y] = cnt
            z +=1


    while y>0:
        cnt -= 1
        y -= 1
        result[x][y] = cnt
        
    output.append(result)

# Output part
for o in range(len(output)):
    print(f'#{o+1}')
    for i in range(len(output[o])):
        for j in range(len(output[o][i])):
            print(int(output[o][i][j]), end = ' ')
        print()
````
